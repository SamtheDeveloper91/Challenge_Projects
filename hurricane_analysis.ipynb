```python
# write your update damages function here to convert values from M or B into actual dollars:
def updated_dollars(amounts_list):
    amounts_list_clean = []
    for i in amounts_list:
        if 'M' in i and '.' in i:
            amounts_list_clean.append(i.replace('M', '00000'))
        elif 'M' in i and not '.' in i:
            amounts_list_clean.append(i.replace('M', '000000'))
        elif 'B' in i and '.' in i:
            amounts_list_clean.append(i.replace('B', '00000000'))
        elif 'B' in i and not '.' in i:
            amounts_list_clean.append(i.replace('B', '000000000'))
        else:
            amounts_list_clean.append(i)
    for i in range(len(amounts_list_clean)):
        if '.' in amounts_list_clean[i]:
            amounts_list_clean[i] = amounts_list_clean[i].replace('.', '')
        else:
            continue
    return amounts_list_clean
#for pair in zip(damages, updated_dollars(damages)):
 #   print(pair)
```

```python
# write your count affected areas function here:
def area_count(areas_affected):
    unique_areas = []
    counts = []
    for area_list in areas_affected:
        for area in area_list:
            if area not in unique_areas:
                unique_areas.append(area)
            else:
                continue
    return unique_areas
```

```python
#define the keys of the affected areas function as the unique Atlantic locations
unique_areas = area_count(areas_affected_clean)
```

```python
#define function for pairing the counts of each area with the keys in unique_areas
def area_counts(areas_affected_clean):
    unique_areas = area_count(areas_affected_clean)
    counts = []
    for i in range(len(unique_areas)):
        counter = 0
        for j in range(len(areas_affected_clean)):
            for k in range(len(areas_affected_clean[j])):
                if areas_affected_clean[j][k] == unique_areas[i]:
                    counter += 1
        counts.append(counter)
    area_counts = {key: value for key, value in zip(unique_areas, counts)}
    return area_counts
```

```python
#show the area_counts dictionary results!
area_hurricane_counts = area_counts(areas_affected_clean)
```

```python
# write your find most affected area function here:
def max_count(area_hurricane_counts):
    top_places = []
    max_location_count = 0
    for i in area_hurricane_counts:
        if area_hurricane_counts[i] >= max_location_count:
            max_location_count = area_hurricane_counts[i]
    for i in area_hurricane_counts:
        if area_hurricane_counts[i] == max_location_count:
            top_places.append(i)
    return max_location_count, top_places
top_cane_place = max_count(area_hurricane_counts)
top_cane_place
```

```python
# write your greatest number of deaths function here:
def greatest_deaths(cane_dict):
    max_deaths = 0
    for k in cane_dict.keys():
        if cane_dict[k]['Deaths'] > max_deaths:
            max_deaths = cane_dict[k]['Deaths']
            max_deaths_name = k
    return max_deaths_name, max_deaths
    
greatest_deaths(hurricane_dict_by_name)
```

```python
#create mortality scale dictionary
mortality_scale = {0: 0,
                   1: 100,
                   2: 500,
                   3: 1000,
                   4: 10000}
```

```python
# write your catgeorize by mortality function here:
def death_rating(cane_dict):
    death_rating = 0
    for key in cane_dict.keys():
        for i in range(len(mortality_scale)):
            if cane_dict[key]['Deaths'] > mortality_scale[i]:
                death_rating = i
        cane_dict[key]['Death Rating'] = death_rating
    return cane_dict

#print(hurricane_dict_by_name['Mitch']['Deaths'])
#print(death_rating)
```

```python
# write your greatest damage function here:
# using similar structure as the greatest deaths function
def greatest_damage(cane_dict):
    damages_dict = {key: int(cane_dict[key]['Damage']) for key in cane_dict.keys() if cane_dict[key]['Damage'] != 'Damages not recorded'}
    max_damage = 0
    max_damage_names = []
    for damage in damages_dict.values():
        if damage > max_damage:
            max_damage = damage
    for key in damages_dict.keys():
        if damages_dict[key] == max_damage:
            max_damage_names.append(key)
    if len(max_damage_names) == 1:
        max_damage_names = max_damage_names[0]
    return max_damage, max_damage_names
    
greatest_damage(hurricane_dict_by_name)
```
